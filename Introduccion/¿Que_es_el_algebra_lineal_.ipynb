# ¿Que es el algebra lineal?
El álgebra lineal es una rama de las matemáticas que estudia los vectores, los espacios vectoriales, las transformaciones lineales y los sistemas de ecuaciones lineales. En esencia, se enfoca en el estudio de las relaciones lineales y sus propiedades, utilizando un enfoque que puede ser visualizado como la manipulación de puntos y líneas en un espacio multidimensional.

### Importamos las librerias necesarias.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
vector_A_3D = np.array([2, 3, 4])
vector_B_3D = np.array([3, -1, 2])
escalar = 1.5

# --- CÁLCULOS EN 3D ---
suma_3D = vector_A_3D + vector_B_3D
escalado_3D = escalar * vector_A_3D

# --- FUNCIÓN PARA DIBUJAR VECTORES 3D ---
def plot_vector_3d(ax, vector, color, label):
    """Dibuja un vector 3D desde el origen."""
    ax.quiver(0, 0, 0, vector[0], vector[1], vector[2],
              color=color, length=1, arrow_length_ratio=0.1, label=label)

# --- CREAR LA FIGURA Y EL SUBPLOT 3D ---
fig = plt.figure(figsize=(14, 7))

# Subplot 1: Representación 3D de la suma y escalamiento
ax1 = fig.add_subplot(121, projection='3d') # 1 fila, 2 columnas, primer gráfico

ax1.set_title('Espacio Vectorial 3D: Suma y Escalamiento', fontsize=14)

# Dibujar los vectores originales
plot_vector_3d(ax1, vector_A_3D, 'blue', f'Vector A = {vector_A_3D}')
plot_vector_3d(ax1, vector_B_3D, 'green', f'Vector B = {vector_B_3D}')

# Dibujar el vector suma
plot_vector_3d(ax1, suma_3D, 'red', f'Suma A+B = {suma_3D}')

# Dibujar el vector escalado
plot_vector_3d(ax1, escalado_3D, 'purple', f'Escalado {escalar}*A = {escalado_3D}')

# Configuración de los ejes 3D
ax1.set_xlim([-5, 5])
ax1.set_ylim([-5, 5])
ax1.set_zlim([-5, 5])
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')
ax1.set_zlabel('Eje Z')
ax1.legend()
ax1.grid(True)
ax1.view_init(elev=20, azim=45) # Ángulo de visión para mejor perspectiva

# Subplot 2: Ilustración conceptual de "Dimensiones" o "Componentes"
ax2 = fig.add_subplot(122) # 1 fila, 2 columnas, segundo gráfico

ax2.set_title('Concepto de Componentes en un Espacio N-Dimensional', fontsize=14)
ax2.text(0.5, 0.9, 'Aunque no podemos graficar N>3 dimensiones,',
         horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes, fontsize=10)
ax2.text(0.5, 0.85, 'cada vector sigue siendo una combinación lineal de sus componentes.',
         horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes, fontsize=10)

# Simulación de un vector de alta dimensión (no se grafica, solo se representa)
vector_ND = np.array([1, 2, 0.5, 3, -1, 2.5, 4, 0.8, -0.3, 1.7])
ax2.text(0.5, 0.7, f'Vector N-D (ej. 10 dimensiones):\n{vector_ND}',
         horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes,
         bbox=dict(boxstyle="round,pad=0.5", fc="cyan", ec="b", lw=1, alpha=0.5), fontsize=10)

ax2.text(0.5, 0.45, 'Las operaciones (suma, escalamiento) se realizan componente a componente',
         horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes, fontsize=10)
ax2.text(0.5, 0.35, 'y el resultado es siempre un vector en el mismo espacio N-D.',
         horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes, fontsize=10)

ax2.text(0.5, 0.2, 'Este es el núcleo del Álgebra Lineal en cualquier número de dimensiones.',
         horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes,
         fontweight='bold', fontsize=12, color='darkgreen')

ax2.axis('off') # Ocultar los ejes para este subplot conceptual

plt.tight_layout(rect=[0, 0, 1, 0.95]) # Ajustar el layout y dejar espacio para el título general
plt.suptitle('Demostración de Espacios Vectoriales Multidimensionales en Álgebra Lineal', fontsize=16, y=0.99)
plt.savefig("espacio_vectorial_multidimensional.png")
plt.show()
```

# grafico de ejemplo de resolucion de incognita de ecuacion de algebra lineal.

#### Dos amigos desiden apostar una carrera, Marcelo y Diego, Diego corre a 3 metros popr segundo, mientras Marcelo corre a 6 metros por segundo, es por esto que Marcelo decide darle una ventaja de 20 segundos. ¿Cuanto tiempo tardara Marcelo en alcanzar a Diego? y ¿Cuata distancia habran recorrido hasta ese punto?

```python
t = np.linspace(0, 60, 1000) # inicio, final, n puntos
```

### Distancia recorrida por Diego

```python
D = 3 * t
```

### Distancia recorrida por Marcelo.

```python
M = 6 * (t-20)
```

### Graficando el punto exacto donde Marcelo alcanza a Diego.

```python
fig, ax = plt.subplots()
plt.title('Carrera')
plt.xlabel('tiempo (en segundos)')
plt.ylabel('distancia (en metros)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 150])
ax.plot(t, D, c='green')
ax.plot(t, M, c='blue')
plt.axvline(x=40, color='red', linestyle='--')
_ = plt.axhline(y=120, color='red', linestyle='--')
```

# Los posibles 3 Escenarios en el algebra lineal

En álgebra lineal, cuando trabajamos con sistemas de ecuaciones lineales, generalmente nos encontramos con tres escenarios posibles

## 1. Solución Única
Este escenario ocurre cuando las ecuaciones del sistema se intersecan en un solo punto. Imagina dos líneas en un gráfico que se cruzan en un lugar específico. Ese punto de intersección es la única solución para ambas ecuaciones como en el ejemplo anterior.



```python
fig, ax = plt.subplots()
plt.title('Carrera')
plt.xlabel('tiempo (en segundos)')
plt.ylabel('distancia (en metros)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 150])
ax.plot(t, D, c='green')
ax.plot(t, M, c='blue')
plt.axvline(x=40, color='red', linestyle='--')
_ = plt.axhline(y=120, color='red', linestyle='--')
```

## 2. Infinitas Soluciones
En este caso, las ecuaciones son, en esencia, la misma línea (o planos que se superponen completamente en dimensiones superiores). Esto significa que cualquier punto en esa línea es una solución al sistema, y como una línea tiene infinitos puntos, hay infinitas soluciones.

Por ejemplo en el ejercicio que estamos trabajando, imagina que Marcelo y Diego salen al mismo tiempo y con la misma velocidad, en este caso ambos estarian hombro a hombro todo el tiempo, lo que nos daria infinitas soluciones.

### Distancia Recorrida por Diego

```python
D = 3 * t
```

### Distancia Recorrida por Marcelo

```python
M = 3 * t
```

## Graficando el punto exacto donde Marcelo alcanza a Diego.

```python
fig, ax = plt.subplots()
plt.title('Carrera')
plt.xlabel('tiempo (en segundos)')
plt.ylabel('distancia (en metros)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 150])
ax.plot(t, D, c='green')
ax.plot(t, M, c='blue')
```

## 3. Sin Solución
Este escenario ocurre cuando las ecuaciones representan líneas (o planos) que son paralelas y nunca se intersecan. Como no hay puntos en común, no hay ninguna solución que satisfaga todas las ecuaciones del sistema.

Igualmente en el ejercicio que venimos trabajando, ahora imagina que Diego le da la ventaja de 20 segundos a Marcelo, pero ambos corren a la misma velocidad, como resultado Diego nunca va a alcanzar a Marcelo por mas que corran indefinidamente, es decir que no habria solucion para este conjunto de ecuaciones.

### Distancia Recorrida por Diego

```python
D = 3 * t
```

### Distancia Recorrida por Marcelo

```python
M = 3 * (t-20)
```

## Graficando el punto exacto donde Marcelo alcanza a Diego.

```python
fig, ax = plt.subplots()
plt.title('Carrera')
plt.xlabel('tiempo (en segundos)')
plt.ylabel('distancia (en metros)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 150])
ax.plot(t, D, c='green')
ax.plot(t, M, c='blue')
```
